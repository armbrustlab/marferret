```python
# read in data

# filepath = '../data/example_data/Chrsp_1_GeneCatalog_proteins_20180813.PFAM34.0.domtblout.tab'
filepath = '../data/example_data/ThTSP-01_Minidiscusspinulatus-Trinity.6tr.PFAM34.0.domtblout.tab'
df = pd.read_csv(
    filepath, 
    engine='python', 
    sep='\s+', 
    skiprows=[0, 1, 2], 
    skipfooter=10, 
    usecols=[0, 3, 4, 6, 7],
    names=['aa_id', 'pfam_name', 'pfam_id', 'pfam_eval', 'pfam_score']
)

df
```

```python
# take the best result for each aa_id

best_annot_df = df.iloc[df.groupby(['aa_id']).pfam_score.idxmax()]
best_annot_df.reset_index(drop=True, inplace=True)

# think about tiebreaker?

best_annot_df
```

```python
# if 6 frame translations were conducted, take the best annotation from the 6 frames

best_annot_df['nt_id'] = best_annot_df['aa_id'].str.slice(start=0, stop=-2)
best_annot_df['frame'] = best_annot_df['aa_id'].str.slice(start=-1)

best_annot_df
```

```python
# select best frame annotation

output_df = best_annot_df.iloc[best_annot_df.groupby(['nt_id']).pfam_score.idxmax()]
output_df.reset_index(drop=True, inplace=True)

output_df
```

```python
# save output file

output_df.to_csv()
```
